```python
import pandas as pd
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
starbucks_data=pd.read_csv("../data/starbucks.csv")
starbucks_data
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
starbucks_data.iloc[::40,[3,11,12]]
```

```python
# select all rows with more than and including 400 calories
starbucks_data[starbucks_data["calories"]>=400]
```

```python
# create a new column containing the caffeine per calories of each drink
starbucks_data["caffeine per calories"]=starbucks_data["caffeine"]/starbucks_data["calories"]
```

```python
# select all rows whose vitamin c content is higher than the iron content
starbucks_data[starbucks_data["vitamin c"]>starbucks_data["iron"]]
```

```python
# what is the average calorie across all items?
average_calorie=starbucks_data["calories"].mean()
average_calorie
```

```python
# how many different categories of beverages are there?
print(starbucks_data["beverage"].unique())
```

```python
# what is the average # calories for each beverage category?
beverages=starbucks_data.groupby("beverage")
print(beverages["calories"].mean())
```

```python
# plot the distribution of the number of calories in drinks with a histogram
plot1=starbucks_data["calories"].plot.hist(title="Calories in Starbucks Drinks")
plot1.set_xlabel("Calories per Drink")
plot1.set_ylabel("Number of Drinks")
```

```python
# plot calories against total fat with a scatterplot
starbucks_data.plot.scatter(x="calories", y="total fat", title="Calories vs Total Fat")
```
